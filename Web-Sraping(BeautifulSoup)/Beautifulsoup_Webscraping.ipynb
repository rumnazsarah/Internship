```python
#2. Write a python program to display IMDB's Top rated 100 movies' data(i.e. name,rating,year of release) and make data frame.
```

```python
url1=get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
url2=get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt")
```

```python
request1=url1.text
request2=url2.text
```

```python
from bs4 import BeautifulSoup as Soup
```

```python
soup_data1=Soup(request1,'html.parser')
soup_data2=Soup(request2,'html.parser')
```

```python
movies1=soup_data1.findAll('div', {'class':'lister-item mode-advanced'})
movies2=soup_data2.findAll('div', {'class':'lister-item mode-advanced'})
```

```python
Name=[]
Rating=[]
Year_of_Release=[]

for x in movies1:
    Name.append(x.h3.a.text)
    Rating.append(x.find('div',{'class':'inline-block ratings-imdb-rating'})['data-value'])
    Year_of_Release.append(x.find('span',{'class':'lister-item-year text-muted unbold'}).text[1:5])
    
```

```python
data=list(zip(Name,Rating,Year_of_Release))
```

```python
import pandas as pd
df1=pd.DataFrame(data,columns=['Name','Rating','Year of Release'])
```

```python
df1
```

```python
url5=get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

```python
request=url5.text   
```

```python
soup5=Soup(request,'html.parser')
```

```python
request
```

```python
odifull=soup5.find_all('tr',{'class':'table-body'})
```

```python
Position=[]
Team=[]
Matches=[]
Points=[]
Rating=[]

for i in odifull:
    Position.append(i.find('td',{'class':'table-body__cell table-body__cell--position u-text-right'}))
    Team.append(i.find('span',{'class':'u-hide-phablet'}))
    Matches.append(i.find('td',{'class':'table-body__cell u-center-text'}))
    Points.append(i.find('td',{'class':'table-body__cell u-center-text'}))
    Rating.append(i.find('td',{'class':'table-body__cell u-text-right rating'}))
```

```python
import pandas as pd
```

```python
data_ODI=list(zip(Position,Team,Matches,Points,Rating))
```

```python
df_ODI=pd.DataFrame(data_ODI,columns=['Position','Team','Matches','Points','Rating'])
```
